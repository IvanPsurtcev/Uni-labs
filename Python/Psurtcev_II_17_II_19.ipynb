# Задание II.19


```python
x = np.arange(-0.5, 0.5, 0.001)
y = np.arange(-0.5, 0.5, 0.001)
x,y = np.meshgrid(x, y)

z1 = np.sqrt((1 + (x**2 + y**2)*2 + np.sqrt(1 + (x**2 + y**2)*4))/2/(x**2 + y**2)/(x**2 + y**2)) - 10
z2 = np.sqrt((1 + (x**2 + y**2)*2 + np.sqrt(1 + (x**2 + y**2)*4))/2/(x**2 + y**2)/(x**2 + y**2)) - 100

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 10)) 

ax1.contour(x, y, z1)
ax2.contour(x, y, z2)

plt.show()
```
